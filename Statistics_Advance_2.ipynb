```python
#1) What is hypothesis in statistics.
Ans) Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data.
It involves formulating two competing statements.
the null hypothesis (H₀), which represents the status quo or no effect, and the alternative hypothesis (H₁), which represents a new claim or effect.
Using sample data, we calculate a test statistic and compare it to a critical value or p-value to determine whether to reject H₀.

#2) What is null hypothesis, and how does it differ from alternative hypothesis.
Null Hypothesis (H₀): A statement that assumes no effect, no difference, or the current belief (e.g., the mean of a population is equal to a certain value).
Alternative Hypothesis (H₁): A statement that contradicts H₀, representing the effect or difference we aim to prove (e.g., the mean is greater than, less than, or not equal to a certain value).
Difference: H₀ is tested for possible rejection, while H₁ is accepted if evidence strongly contradicts H₀.

#3) Explain the significant level in hypothesis testing and its role in deciding the outcome of a test.
Ans) The significance level (α) is the probability of rejecting the null hypothesis when it is true (Type I error). Common choices are 0.05 (5%) or 0.01 (1%).
If p-value ≤ α, we reject H₀ (evidence supports H₁).
If p-value > α, we fail to reject H₀ (insufficient evidence against H₀).
Thus, α controls the threshold for making statistical decisions.

#4) What are type 1 and type 2 errors? give examples of each.
Ans) Type I Error (False Positive): Rejecting H₀ when it is actually true.
Example: Concluding a new drug works when it actually doesn’t.
Type II Error (False Negative): Failing to reject H₀ when it is false.
Example: Concluding a new drug doesn’t work when it actually does.

#5)What is the difference between a Z-test and a T-test? Explain when to use each.
Ans)
Z-test: Used when population variance is known and/or the sample size is large (n > 30). It relies on the standard normal distribution.
T-test: Used when population variance is unknown and the sample size is small (n ≤ 30). It relies on the Student’s t-distribution.
```

```python
#6)
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n, p, size = 10, 0.5, 1000
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(0, n+2), edgecolor='black', density=True)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()
```

```python
#7)
import numpy as np
from scipy import stats

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesis: H0: mu = 50
mu0 = 50
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Z statistic
z_stat = (sample_mean - mu0) / (sample_std / np.sqrt(n))

# p-value (two-tailed)
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

z_stat, p_value
```

```python
#8)
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Simulate normal data
data = np.random.normal(loc=100, scale=15, size=1000)

# 95% Confidence Interval
mean = np.mean(data)
std_err = stats.sem(data)
ci = stats.norm.interval(0.95, loc=mean, scale=std_err)

# Plot
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(ci[0], color='red', linestyle='--', label=f"Lower CI: {ci[0]:.2f}")
plt.axvline(ci[1], color='green', linestyle='--', label=f"Upper CI: {ci[1]:.2f}")
plt.axvline(mean, color='blue', linestyle='-', label=f"Mean: {mean:.2f}")
plt.title("Normal Distribution with 95% CI")
plt.legend()
plt.show()

ci
```
